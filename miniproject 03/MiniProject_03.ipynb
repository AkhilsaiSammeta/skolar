2. Code Execution Plan

a. K-Means Clustering Example

1 Import necessary libraries:

```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
```

2 Load the dataset and apply K-Means:

```python
data = load_iris()
X = data.data

kmeans = KMeans(n_clusters=3)
kmeans.fit(X)
labels = kmeans.labels_

# Plot results
plt.scatter(X[:, 0], X[:, 1], c=labels)
plt.title('K-Means Clustering')
plt.show()
```

3 Use Elbow Method to find optimal clusters:

```python
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)

plt.plot(range(1, 11), wcss)
plt.title('Elbow Method')
plt.show()
```

b. Hierarchical Clustering Example

1 Import necessary libraries:

```python
from sklearn.cluster import AgglomerativeClustering
import scipy.cluster.hierarchy as sch
```

2 Plot dendrogram and apply hierarchical clustering:

```python
# Dendrogram
dendrogram = sch.dendrogram(sch.linkage(X, method='ward'))
plt.title('Dendrogram')
plt.show()

# Apply Agglomerative Clustering
hc = AgglomerativeClustering(n_clusters=3)
labels_hc = hc.fit_predict(X)

# Plot clusters
plt.scatter(X[:, 0], X[:, 1], c=labels_hc)
plt.title('Hierarchical Clustering')
plt.show()
```

c. DBSCAN Example
1 Import necessary libraries:

```python
from sklearn.cluster import DBSCAN
```

2 Apply DBSCAN:

```python
dbscan = DBSCAN(eps=0.5, min_samples=5)
labels_dbscan = dbscan.fit_predict(X)

# Plot clusters
plt.scatter(X[:, 0], X[:, 1], c=labels_dbscan)
plt.title('DBSCAN Clustering')
plt.show()
```

d. Principal Component Analysis (PCA) Example

1 Import necessary libraries:

```python
from sklearn.decomposition import PCA
```

2 Apply PCA:

```python
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

plt.scatter(X_pca[:, 0], X_pca[:, 1], c=labels)
plt.title('PCA')
plt.show()
```

e. t-SNE Example

1 Import necessary libraries:

```python
from sklearn.manifold import TSNE
```

2 Apply t-SNE:

```python
tsne = TSNE(n_components=2)
X_tsne = tsne.fit_transform(X)

plt.scatter(X_tsne[:, 0], X_tsne[:, 1], c=labels)
plt.title('t-SNE')
plt.show()
```
